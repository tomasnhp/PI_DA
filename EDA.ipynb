
### Análisis de Variables Categóricas:

- **LUGAR_DEL_HECHO:** Lugares donde ocurren los incidentes.
- **TIPO_DE_CALLE:** Tipos de calles involucradas en los incidentes.
- **Calle:** Nombres de calles donde ocurren los incidentes.
- **Cruce:** Calles de cruce en los incidentes.
- **Dirección Normalizada:** Direcciones normalizadas de los incidentes.
- **XY (CABA):** Coordenadas geográficas en formato de cadena.
- **PARTICIPANTES:** Tipos de participantes en los incidentes.
- **VICTIMA:** Información sobre las víctimas.
- **ACUSADO:** Información sobre los acusados.

```python
# Análisis de Variables Categóricas
top_location = df['LUGAR_DEL_HECHO'].mode()[0]
top_street_type = df['TIPO_DE_CALLE'].mode()[0]
top_street_name = df['Calle'].mode()[0]
top_cross_street = df['Cruce'].mode()[0]
top_normalized_address = df['Dirección Normalizada'].mode()[0]
top_participants = df['PARTICIPANTES'].mode()[0]
top_victim_vehicle = df['VICTIMA'].mode()[0]
top_accused_vehicle = df['ACUSADO'].mode()[0]

print("Análisis de Variables Categóricas:")
print(f"- Lugar más común: {top_location}")
print(f"- Tipo de calle más común: {top_street_type}")
print(f"- Calle más comúnmente involucrada: {top_street_name}")
print(f"- Calle de cruce más común: {top_cross_street}")
print(f"- Dirección normalizada más común: {top_normalized_address}")
print(f"- Tipo de participante más común: {top_participants}")
print(f"- Vehículo más comúnmente involucrado como víctima: {top_victim_vehicle}")
print(f"- Vehículo más comúnmente involucrado como acusado: {top_accused_vehicle}")

# Relaciones de Variables
min_date = df['FECHA'].min()
max_date = df['FECHA'].max()
avg_victims = df['N_VICTIMAS'].mean()
avg_altura = df['Altura'].mean()
comuna_counts = df['COMUNA'].value_counts()
pos_x_present = df['pos x'].count()
pos_y_present = df['pos y'].count()

print("\nRelaciones de Variables:")
print(f"- Fecha mínima: {min_date}")
print(f"- Fecha máxima: {max_date}")
print(f"- Cantidad promedio de víctimas por incidente: {avg_victims:.2f}")
print(f"- Altura promedio donde ocurren los incidentes: {avg_altura:.2f}")
print("\nDistribución de incidentes por comuna:")
print(comuna_counts)
print(f"\nCoordenadas geográficas 'pos x' presentes en todos los incidentes: {pos_x_present == len(df)}")
print(f"Coordenadas geográficas 'pos y' presentes en todos los incidentes: {pos_y_present == len(df)}")
```

## ANALISIS GEO-ESPACIAL DE INCIDENTES

## Clustering espacial

CREACION DE CLUSTERS Y VISUALIZACION

```python
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
from sklearn.cluster import KMeans


# Número de clusters
n_clusters = 5

# Crear un modelo K-Means
kmeans = KMeans(n_clusters=n_clusters, random_state=0)

# Ajustar el modelo a las coordenadas de tus datos
coordinates = gdf[['pos x', 'pos y']].values
gdf['cluster'] = kmeans.fit_predict(coordinates)

# Etiquetas de los clusters
cluster_labels = ['Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4', 'Cluster 5']

# Asignar etiquetas a los clusters
gdf['cluster'] = gdf['cluster'].map({i: label for i, label in enumerate(cluster_labels)})

# Crear una figura y ejes para el gráfico
fig, ax = plt.subplots(figsize=(12, 8))

# Visualizar los clusters
gdf.plot(column='cluster', cmap='Set1', legend=True, markersize=50, ax=ax)
plt.title('Clustering Espacial de Incidentes')
plt.axis('off')  # Desactivar ejes para un aspecto más limpio

# Mostrar el gráfico
plt.show()

# Variables categóricas que deseas analizar
cat_vars = ['TIPO_DE_CALLE', 'COMUNA', 'VICTIMA', 'ACUSADO']  # Puedes ajustar esta lista según tus necesidades


# Calcular la distribución de variables categóricas en cada cluster
for var in cat_vars:
    cluster_counts = gdf.groupby(['cluster', var]).size().unstack(fill_value=0)
    cluster_counts = cluster_counts.div(cluster_counts.sum(axis=1), axis=0)
    cluster_stats = cluster_stats.join(cluster_counts.add_prefix(f'{var}_'))

# Mostrar las estadísticas de los clusters
print(cluster_stats)
```

DISTRIBUCION DE INCIDENTES POR CLUSTER

```python
# Contar cuántos incidentes hay en cada cluster
incidentes_por_cluster = gdf['cluster'].value_counts().reset_index()
incidentes_por_cluster.columns = ['Cluster', 'Número de Incidentes']

# Crear un gráfico de barras para visualizar la distribución
plt.figure(figsize=(10, 6))
plt.bar(incidentes_por_cluster['Cluster'], incidentes_por_cluster['Número de Incidentes'], color='skyblue')
plt.xlabel('Cluster')
plt.ylabel('Número de Incidentes')
plt.title('Distribución de Incidentes por Cluster')
plt.xticks(rotation=45)
plt.show()
```
